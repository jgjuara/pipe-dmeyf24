```python
import lightgbm as lgb
import os
import joblib
```

```python
# Load the model from a .txt file
model_path = 'lgbm-lgbm_binaria_t3-21-549937.txt'
```

```python
os.path.exists(model_path)
```

```python
# read string from file
with open(model_path, 'r') as file:
    model_str = file.read()
```

```python
gbm_txt = lgb.Booster(model_str=model_str)
```

```python
lgb.plot_importance(gbm_txt, importance_type='split', max_num_features=100, figsize=(10, 15))
```

```python
gbm_pickle = joblib.load(model_path)
```

```python
lgb.plot_importance(gbm_pickle, importance_type='split', max_num_features=100, figsize=(10, 15))
```
